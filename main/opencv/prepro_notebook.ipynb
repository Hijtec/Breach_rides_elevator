```python
import cv2
import preprocess
import imutils
from imutils import contours
from matplotlib import pyplot as plt
image = cv2.imread(r'E:\GoogleDriveSync\BREACH_RIDES_ELEVATOR\breach_rides_elevator\main\opencv\data\test_data\a2.png', cv2.COLOR_BGR2RGB)
```

```python
image = preprocess.adjust_brightness_dynamic(image)

gray = preprocess.grayscale(image)
color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(gray, cmap="gray")
plt.title('my picture')
plt.show()
```

```python
median_filtered = preprocess.median_filter(gray)
plt.imshow(median_filtered, cmap="gray")
plt.title('filtered')
plt.show()
```

```python
edges = preprocess.canny_edge_extraction(median_filtered, sigma = 0.5,dilate_iterations=2,erode_iterations=1)
plt.imshow(edges)
plt.title('canny edges')
plt.show()
```

```python
orig_label, sort_label, cnts, cnts_ordered = preprocess.candidate_extraction(median_filtered, edges)
plt.imshow(sort_label)
plt.title('labeled_sorted')
plt.show()
plt.imshow(orig_label)
plt.title('labeled')
plt.show()
```

```python
boxed_contours, boxes = preprocess.box_contours(median_filtered, cnts)
plt.imshow(boxed_contours)
plt.title('contours boxed')
plt.show()
```

```python
median_filtered = preprocess.median_filter(gray)
filtered, boxes_filtered = preprocess.boxdimension_filter(median_filtered, boxes, size_lim=(300,80000), width_lim=(10,200), height_lim=(10,200), ratio_wh_lim=(0.2,4))
plt.imshow(filtered)
plt.title('filtered bounding boxes')
plt.show()
```
